 # 2011

```python
df_2011 = pd.read_csv("11zpallnoagi.csv", dtype={'ZIPCODE':'object'})
```

```python
df_2011.drop(['STATEFIPS'], axis=1, inplace=True)
```

```python
df_2011.drop(['AGI_STUB'], axis=1, inplace=True)
```

```python
df_2011.columns
```

```python
df_2011.rename(columns={'N1':'num_returns',
                        'MARS2':'joint_returns',
                       'A00100':'adjust_gross_income',
                       'A00200':'salaries_wages',
                       'A00300':'taxable_interest',
                       'A00600':'dividends',
                       'A00650':'qualified_dividends',
                       'A18500':'real_estate_taxes',
                       'A18300':'total_taxes',
                       'A04800':'taxable_income_amount',
                       'N07260':'num_returns_residential_energy_credits',
                       'A07260':'residential_energy_tax_credit',
                       'A06500':'income_tax_amount',
                       'A10300':'total_tax_liability'}, inplace=True)
```

```python
zips = df_2011.groupby('ZIPCODE').median()
zips['year'] = 2011
```

```python
zips.reset_index(inplace=True)
```

```python
zips[(zips.ZIPCODE == '10510') | (zips.ZIPCODE == '01010')]
```

```python
# pd.set_option('max_rows', 100)
# df_2011.dtypes
```

```python
zips.drop([0, 27788], inplace=True)
```

```python
 with open("2011_zipcodes.pkl", "wb") as f:
    pickle.dump(zips, f)
```
